```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
iris = pd.read_csv("IRIS.csv")
```

```python
print(iris.head())
```

```python
print(iris.describe())
```

```python
print("Target Labels", iris["species"].unique())
```

```python
# Define a dictionary to map species to colors
species_to_color = {'Iris-setosa': 'red', 'Iris-versicolor': 'green', 'Iris-virginica': 'blue'}

# Create a figure and a set of subplots
fig, ax = plt.subplots()

# Plot the data
for species in iris["species"].unique():
    species_data = iris[iris["species"] == species]
    ax.scatter(species_data["sepal_width"], species_data["sepal_length"], label=species, color=species_to_color[species])

# Set the labels for the x and y axes
ax.set_xlabel("Sepal Width")
ax.set_ylabel("Sepal Length")

# Set the title of the plot
ax.set_title("Iris Sepal Width vs. Sepal Length")

# Add a legend
ax.legend()

# Display the plot within the notebook
plt.show()
```

```python
# Dropping the column "species" from the dataset and storing the remaining data in x
x = iris.drop("species", axis=1)

# Storing the column "species" in y
y = iris["species"]

# Splitting the dataset into training and testing sets
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, 
                                                    test_size=0.2, 
                                                    random_state=0)

# Importing KNeighborsClassifier and creating an instance knn
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1)

# Fitting the model using the training data
knn.fit(x_train, y_train)
```

```python
# x_new = np.array([[5, 2.9, 1, 0.2]])
# prediction = knn.predict(x_new)
# print("Prediction: {}".format(prediction))

# New observation to predict
x_new = np.array([[5, 2.9, 1, 0.2]])

# Use the trained KNN model to make a prediction
prediction = knn.predict(x_new)

# Print the prediction in a user-friendly format
print(f"Prediction: {prediction[0]}")
```
